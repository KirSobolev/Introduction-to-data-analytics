```python
url = "https://en.wikipedia.org/wiki/Rovaniemi"
# Now wikipedia requires user-agent
headers = {
    "User-Agent": "MyApp/1.0 (your_email@example.com) Python requests"
}
response = requests.get(url, headers=headers)
soup = BeautifulSoup(response.content, "html.parser")
```

```python
# Find the infobox containing coordinates
infobox = soup.find("table", {"class": "infobox"})
# Find span with class geo
geo_span = infobox.find("span", {"class": "geo"})
# Parse coordinates
coordinates = geo_span.text if geo_span else None

print("Coordinates:", coordinates)
```

```python
# Inside infobox find div with nickname class
ib_settlement_nickname = infobox.find("div", {"class": "ib-settlement-nickname"})
# Parse the text
nickname = ib_settlement_nickname.text
print(f"Nickname(s): {nickname}")
```

```python
# Inside infobox find div with nickname class
infobox_data = infobox.find_all("td", {"class": "infobox-data"})
# Using index is a hardcoded solution which will probably break at some point
population = infobox_data[10].text
print(f"Population: {population}")
```

```python
# Search the whole table fot "Total" - the first one will be total area, second total population, but again not optimal solution
for row in infobox.find_all("tr"):
    header = row.find("th")
    value = row.find("td")
    if header and "Total" in header.text:
        total_area = value.text.strip()
        break
```

```python
# Remove commas, extract first number
total_area = re.findall(r'[\d,.]+', total_area)[0]
total_area = float(total_area.replace(',', ''))
print(f"Total area: {total_area}")
```

Unemployment table

```python
url = "https://en.wikipedia.org/wiki/List_of_countries_by_unemployment_rate"
headers = {
    "User-Agent": "MyApp/1.0 (your_email@example.com) Python requests"
}
response = requests.get(url, headers=headers)
soup = BeautifulSoup(response.content, "html.parser")

# Find the table with the class 'wikitable'
unemployment = soup.find("table", {"class": "wikitable"})

# Parse the table into a DataFrame and delete old index
df_unemployment = pd.read_html(str(unemployment))[0]
# Rename temperature column
df_unemployment
```

```python
# Handle year in unemployment table
df_unemployment['Year'] = df_unemployment['Source/date of information'].str.extract(r'(\d{4})').astype(int)
# Keep only 2020+
df_unemployment = df_unemployment[df_unemployment.Year >= 2020].drop(columns="Source/date of information")
# Rename Country column
df_unemployment = df_unemployment.rename({"Country/Territory": "Country"}, axis=1)
# Strip * from country names
df_unemployment['Country'] = df_unemployment['Country'].str.rstrip('*').str.strip()
df_employment['Country'] = df_employment['Country'].str.rstrip('*').str.strip()
# Handle Year in employment
df_employment["Year"] = df_employment["Year"].str.extract(r'(\d{4})').astype(int)
# Keep only 2020+ and drop ages with source
df_employment = df_employment[df_employment.Year >= 2020].drop(columns=["Ages", "Source"])
# Rename employment rate
df_employment = df_employment.rename({"Rate .mw-parser-output .nobold{font-weight:normal}(%)": "Employment rate (%)"}, axis=1)
```

What we need to do is access every rows price, count them together and divide by amount of rows

<ul>
<li><b>3. Download the "complex.xml" file from Moodle, and create a DataFrame out of it.</b></li>
<ul>
    <li><b>Note:</b> using pandas.read_xml() would otherwise work, but it ignores the "ownership" â€“field. You'll have to find another way to get this data into your DataFrame as well!</li>
</ul>
</ul>

```python
tree = ET.parse("complex.xml")
root = tree.getroot()

rows = []
# Search each element
for car in root.findall("car"):
    # Parse usual data
    row = {
        "id": int(car.find("id").text),
        "brand": car.find("brand").text,
        "model": car.find("model").text,
        "year": int(car.find("year").text),
        "price": float(car.find("price").text),
    }
    
    # Flatten ownership
    ownership = car.find("ownership")
    if ownership is not None:
        row["ownership_company"] = ownership.findtext("company")
        row["ownership_acquired"] = ownership.findtext("acquired")
        payment = ownership.find("payment_info")
        if payment is not None:
            row["ownership_credit_card"] = payment.findtext("credit_card")
            row["ownership_iban"] = payment.findtext("iban")
    
    rows.append(row)

df = pd.DataFrame(rows)
df
```
